```python
class Laptop:
    def __init__(self, brand, ram_gb, battery=100):
        self.brand = brand
        self.ram_gb = ram_gb
        self.battery = battery

    def run_task(self, minutes):
        # Reduce battery by (minutes * 1.5). If battery <= 0, print "Shutting down" and set to 0
        # Else print remaining battery
        self.battery -= minutes * 1.5 # First reduce the battery
        if self.battery <= 0:
          print("Shutting down")
          self.battery = 0
        else:
          print(f"Battery Remaining: {self.battery}")
        pass

    def charge(self, hours):
        # Increase battery by (hours * 20). Cap at 100
        # Print new battery level
        self.battery+=hours*0.20
        if self.battery>100:
          self.battery=100
        print(f"Battery Remaining: {self.battery}")
        pass


#Make and call the objects:

Lapt1 = Laptop("Dell", 8, 100)
Lapt2 = Laptop("Apple", 16, 100)

Lapt1.run_task(10)
Lapt1.run_task(30)
Lapt1.charge(2)
Lapt1.run_task(100)
```
