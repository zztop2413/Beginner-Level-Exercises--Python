```python
from collections import Counter
import string

def read_file(filename):
  with open(filename, "r") as file:
    file=file.read()

    #Convert all to lowercase
    file=file.lower()

    #Avoid punctuation marks:
    for i in string.punctuation:
      file=file.replace("i","")

    #Split each word from file
    file=file.split()

    #Use Counter method from collections to count the number of words with highest frequency
    highest_word_count = Counter(file)
    highest_word_count=max(highest_word_count, key=highest_word_count.get)

    #Print highest frequent word
    print(highest_word_count)

read_file("/content/sample_data/data.txt")
```

```python
# Get number as integer
num = int(input("Enter number: "))

# Store original number
original = num

# Reverse the number mathematically
reverse = 0
while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

# Check result
if original == reverse:
    print("Mirror Number")
else:
    print("Not Mirror Number")
```
