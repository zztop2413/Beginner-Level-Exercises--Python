**Practice Problem 1**

Get a sample text from sample.txt and count the following:




*   No of Lines:
*   No of Words:
*   No of Characters:

and save it to counter_of_string.txt file

```python
import string
from functools import reduce

with open("./sample_data/sample.txt","r") as file:
  #----------------------------------No of lines-----------------------------------------------:
  line_count= sum(1 for line in file)
  print(f"No of Lines: {line_count}")

  file.seek(0)
  #----------------------------------No of words (2)-------------------------------------------:

 # 1. Character Count:

  # Total Word Count:
  currently_inside_word=0
  total_words=0

  for char in file.read():
    # STEP A: Check if we are hitting a separator (Space or Punctuation)
    if char.isspace() or (char in string.punctuation and char != "'"):
      currently_inside_word = False # Flip switch Off

          # STEP B: Check if we just started a new word
    elif currently_inside_word == False:
      total_words += 1           # Count the new word!
      currently_inside_word = True  # Flip switch On
  print(f"No of words using State Machine Me: {total_words}")
  file.seek(0) # Bring cursor back to index 0


  #2. Word Count using lamda() function:
  strings_list=file.readlines()
  words_per_line=list( map(lambda x: len(x.split() ), strings_list) )
  print(words_per_line) # Print each individual words

  total_words=reduce(lambda x,y: x+y, words_per_line)
  print(f"No of words using split() method: {total_words}")

  #----------------------------------No of characters----------------------------------------:
  #First, space and newline characters have to be omitted

  print(strings_list) #Printing before space and newlines have been ommited

  #New line deleted
  strings_list=list(map(str.strip, strings_list))

  #Remove whitespace
  strings_list=list(map(lambda x: x.replace(" ",""), strings_list ))

  print(strings_list) #Printing after space and newlines have been ommited

  #No of Characters:
  no_of_characters= list(map(lambda x: len(x), strings_list))
  no_of_characters=reduce(lambda x,y: x+y, no_of_characters)

  print(f"No of Characters: {no_of_characters}")
```
