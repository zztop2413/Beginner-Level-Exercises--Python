```python
file=open("./sample_data/sample.txt","r")

# Read the entire file and put in a list
content=file.readlines()
print(content)

#Close the file:
file.close()
print(file.closed)   # It is a must to close files
```

```python
#Open file "with open()" function
with open("./sample_data/sample.txt","r") as file:
  content=file.readlines()
  print(content)

#Doesn't require file closing externally
print(file.closed)
```

```python

with open("./sample_data/sample.txt","r") as file:
  for line in file:
    print(line, end="")

#Doesn't require file closing externally
print(file.closed)
```

```python
#Seek
with open("./sample_data/test.txt","r") as file:
  #file.tell() tells at which index the curson currently is at
  print(file.tell()) #currently at index 0

  print(file.read(5) ) # reads out the 5
  print(file.tell())

  print(file.seek(0)) # Move the index back n forth at 0
```
