# > Data Cleansing in File


Create a text file named emails.txt containing multiple email addresses, some in uppercase and some repeated.
Write a Python program to clean the data by converting all emails to lowercase, removing duplicates, sorting them alphabetically, and saving the cleaned list to a new file named unique_emails.txt

**Write into Text file**

```python
with open('./sample_data/email.txt', 'w+') as file:
    file.write("test@example.com\n")
    file.write("ANOTHER@example.com\n")
    file.write("Test@example.com\n")
    file.write("duplicate@mail.com\n")
    file.write("DuplicATE@mail.com\n")
    file.write("user1@domain.org\n")
    file.write("user2@domain.org\n")
    file.write("USER1@domain.org\n")
    file.write("unique@email.net\n")
    file.write("ANOTHER@example.com\n")
print("email.txt has been populated with sample data.")
```

```python

with open('./sample_data/email.txt', 'r') as infile:
    # Read all lines from the input file
    emails = infile.readlines()

# 1. Lower Case conversion and stripping whitespace/newlines
emails = [i.strip().lower() for i in emails]
print("Lowercase emails:", emails)

# 2. Removing Duplicates using a set
unique_emails_set=set(emails)
print("Unique emails:", unique_emails_set)

# 3. Convert back to a sorted list:
unique_emails=sorted(list(unique_emails_set))
print("Unique Emails sorted: ",unique_emails)


# 4. Save cleaned file to a new one called email_cleaned
with open("./sample_data/email_cleaned.txt","w") as outfile:
   new_file=outfile.writelines("\n".join(unique_emails))
   print("email_cleaned.txt has been created with cleaned emails.")
```

Write a function create_file(filename, content) that creates a new text file with the given filename and writes the content into it. Call it using positional arguments.

📥 Sample Input:
"notes.txt", "Hello, this is my first file."
📤 Sample Output (notes.txt will contain):
Hello, this is my first file.


```python
def create_file(filename, content):
  with open(filename, 'w') as file:
    file.write(content)

create_file("notes.txt","Hello, this is my first file")
```
